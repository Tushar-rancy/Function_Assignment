iteretors : Iteretor is an object that implements iteration protocol and it is consist of below -

***6. Explain the concept of generators in Python and how they are defined.***

***Answer:***

Generators : Generator is special type of function that generate sequence of the values spontanously intead of generating and sorting entire sequence at once in memory.This make it mamory efficient especially working with large datasets.

--> generators are defined using the yield keyword instead of return within a function.


--> When a generator function is called, it returns a generator object (an iterator).


--> Each time the next() function is called on the generator object, the function executes until it encounters the yield statement.


--> yield returns a value to the caller and pauses the function's execution at that point.


--> When next() is called again, the function resumes execution from where it left off.

Example :

    def my_generator():
        """A simple generator that yields numbers from 1 to 3."""
        yield 1
        yield 2
        yield 3

    # Get a generator object
    gen = my_generator()

    # Get the values using next()
    print(next(gen))  # Output: 1
    print(next(gen))  # Output: 2
    print(next(gen))  # Output: 3
    # print(next(gen))  # Raises StopIteration exception

***10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?***

***Answer :***

1 - map()

Purpose: Applies a given function to each item in an iterable and returns an iterator of the results.

2 - reduce()

Purpose: Applies a function cumulatively to the items of an iterable, from left to right, reducing it to a single value.

3 - filter()

Purpose: Creates an iterator of elements from an iterable for which a given function returns True.
